```python
file = open("example.txt", "r")
top_map = list(map(list, file.read().splitlines()))
```

```python
top_map
```

```python
stack = []
zeroes = []
paths = [[{} for _ in range(len(top_map[i]))] for i in range(len(top_map))]
for i in range(len(top_map)):
    for j in range(len(top_map[i])):
        if top_map[i][j] == "9":
            stack.append((i, j))
            paths[i][j] = (i, j)
        if top_map[i][j] == "0":
            zeroes.append((i, j))

directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
curr = 9
new_stack = []
while curr > 0:
    curr -= 1
    print('\n'.join(list(map(str, paths))))
    print()
    while stack:
        i, j = stack.pop()
        for d in directions:
            x, y = i + d[0], j + d[1]
            if 0 <= x < len(top_map) and 0 <= y < len(top_map[x]) and top_map[x][y] == str(curr):
                paths[x][y].add(paths[i][j])
                new_stack.append((x, y))
    stack = new_stack

paths
```
